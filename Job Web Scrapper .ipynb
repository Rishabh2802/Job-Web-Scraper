# Web Scrapping Monstor.com for python developer jobs

```python
import requests
from bs4 import BeautifulSoup
url='https://www.monster.com/jobs/search/?q=Software-Developer&where=Australia'
page=requests.get(url)
soup=BeautifulSoup(page.content,'html.parser')
#After inspecting the site we got to knew that jobs are under the section where id==ResultContainer.
```

```python
#BeautifulSoup allows to find that specific element easily by it's id.
#It will find the container where id=ResultContainer.
results=soup.find(id='ResultsContainer')
```

```python
print(results.prettify())
# By using prettify function in BeautifulSoup will be able to see how tags are nested.
```

```python
job_elems = results.find_all('section', class_='card-content')
# we have observed that all the job related information like job title,location etc. are in the tag where class=card-content.
#So will fetch all the section where class=card-content.
```

```python
for job_ele in job_elems:
    print(job_ele,end='\n'*2)
```

```python
for job_ele in job_elems:
    title_ele=job_ele.find('h2',class_='title')
    company_ele=job_ele.find('div',class_='company')
    location_ele=job_ele.find('div',class_='location')
    print(title_ele)
    #print(company_ele)
    #print(location_ele)
    print()
#In the card-content section there are sub-section in tags like class=title,class=company,class=location.
#Now we'll store title,company and location in different objects
```

```python
for job_ele in job_elems:
    title_ele=job_ele.find('h2',class_='title')
    company_ele=job_ele.find('div',class_='company')
    location_ele=job_ele.find('div',class_='location')
    if None in (title_ele,company_ele,location_ele):
        continue
    print(title_ele.text.strip())
    print(company_ele.text.strip())
    print(location_ele.text.strip())
    print()
    
```

```python
python_jobs=results.find_all('h2',string=lambda text:'python' in text.lower())
print(len(python_jobs))
for job in python_jobs:
    link=job.find('a')['href']
    print(job.text.strip())
    print(f"Apply here: {link}\n")
```
